```python
from models import FixedTargetEGNCA
from argparse import Namespace
import datetime
import torch
import time

# * format the date and time in a readable format
now = datetime.datetime.now()
date_time = now.strftime('%Y-%m-%d@%H-%M-%S')

torch.manual_seed(42)
torch.set_default_dtype(torch.float32)
pool_size = 64

# * create args
args = Namespace()
args.save_to = 'logs'
args.graph = 'line'
args.size = 8
args.model_name = f'test-{args.graph}{args.size}-{date_time}'
args.scale = 1.0
args.coords_dim = 3
args.hidden_dim = 8
args.message_dim = 16
args.has_attention = True
args.device = 'cuda'
args.pool_size = pool_size
args.batch_size = 2
args.epochs = 25_000
args.min_steps = 16
args.max_steps = 32
args.info_rate = 250
args.lr = 1e-3
args.b1 = 0.9
args.b2 = 0.999
args.wd = 1e-5
args.patience_sch = 500
args.factor_sch = 0.5
args.density_rand_edge = 1.0
args.reset_at = 1_000

# * create and train model
fixed_target_egnca = FixedTargetEGNCA(args)
tik = time.time()
fixed_target_egnca.train_model(
    train_verbose=False,
    view_random_graphs=True,
)
tok = time.time()
print ('train time: %d (s)' % (tok - tik))
```
